Defining both events:
- Let
  $$ 
  A
  $$
  be the event that I was invited to the second round of interviews.
- Let
  $$
  B 
  $$
  be the event of how I felt after the interview.

We are asked to find 
$$
P(A \mid B)
$$
the probability that I will get invited to the second round, given that I feel the interview went well.

### Using Bayes' Theorem:

$$
P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}
$$

Where:
 $$
 P(B \mid A) = 0.9 
 $$
                     (90% of those invited feel the interview went well),
 $$
 P(B \mid A^c) = 0.7
 $$
                     (70% of those not invited feel the interview went well),
 $$
 P(A) = \frac{1}{2}
 $$
                     (only half of those shortlisted were invited to the second stage).


````python
Prob_B_giv_A = 0.9      
Prob_B_not_A = 0.7  
Prob_A = 0.5              
Prob_not_A = 1 - Prob_A      


Prob_B = Prob_B_giv_A * Prob_A + Prob_B_not_A * Prob_not_A

Prob_A_giv_B = (Prob_B_giv_A * Prob_A) / Prob_B

Prob_A_giv_B
````

Now, we can calculate 
$$
P(A \mid B)
$$

$$
P(A \mid B) = \frac{0.9 \cdot \frac{1}{2}}{0.8} = 0.5625
$$

Thus, the probability that I will get invited to the second stage, given that I felt the first interview went well, is:

$$
P(A \mid B) = 0.5625
$$


### b

Based on the result from part A, I can see that feeling good about my first interview slightly boosts my chances of being invited to the second stage. The probability increases to 56.25%, compared to the base rate of 50%—so it does help a little.

That said, the improvement isn’t dramatic. While there’s a connection between feeling positive about my interview and actually moving forward, it’s not a strong enough indicator to guarantee success. In short, while having a good feeling might give me some hope, it's not something I should rely on too heavily when trying to predict the outcome.

````python
import requests
import os
import urllib.request
````

````python
api_url = "https://www.xeno-canto.org/api/2/recordings"
````

````python
def show_species(bird_name):
    query_params = {"query": bird_name}
    response = requests.get(api_url, params=query_params)

    if response.status_code != 200:
        print("Error fetching data.")
        return None

    data = response.json()
    bird_recordings = data.get('recordings', [])

    if not bird_recordings:
        print(f"No recordings found for: {bird_name}")
        return None

    unique_species = {rec['en'] for rec in bird_recordings}
    species_list = list(unique_species)

    if not species_list:
        print("No species found.")
        return None

    print("Available bird species:")
    for idx, species in enumerate(species_list, start=1):
        print(f"{idx}. {species}")

    return species_list
````

````python
def download_bird_audio(selected_species):
    query_params = {"query": selected_species}
    response = requests.get(api_url, params=query_params)

    if response.status_code != 200:
        print("Error fetching data.")
        return

    data = response.json()
    bird_recordings = data.get('recordings', [])

    if not bird_recordings:
        print(f"No recordings found for: {selected_species}")
        return

    folder_name = selected_species.replace(" ", "_")
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)

    print(f"Found {len(bird_recordings)} recordings. Downloading...")
    for idx, rec in enumerate(bird_recordings, start=1):
        if "file" in rec:
            audio_url = rec['file']
            if not audio_url.startswith("https://"):
                audio_url = f"https:{audio_url}"
            file_name = os.path.join(folder_name, f"{selected_species.replace(' ', '_')}_{idx}.mp3")

            try:
                print(f"Downloading {audio_url} as {file_name}...")
                urllib.request.urlretrieve(audio_url, file_name)
                print(f"Downloaded {file_name}")
            except Exception as e:
                print(f"Failed to download {audio_url}: {e}")
        else:
            print(f"No audio file found for recording {idx}")

    print(f"Download complete. Files saved in {folder_name} folder.")
````

## Derivation of the Sigmoid Function:

Let
$$
p
$$
be the probability that a certain event occurs.

Let 
$$
1 - p
$$
be the probability that the event does not occur. Then the **odds** of an event occurring is given by function:


$$
\text{odds} = \frac{p}{1 - p}
$$

This represents the ratio of the probability of the event occurring to the probability of it not occurring.

### Step 1: Log-Odds

In logistic regression, we assume that the **log-odds** of the event occurring, denoted by 
$$
\log \left( \frac{p}{1 - p} \right) 
$$
and has a linear relationship with one or more predictor variables. Denoting this linear combination of predictors by 
$$
z
$$
where:

$$
z = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n
$$

given
$$
\beta_1 ....\beta_n
$$
denotes the weights while
$$
x_1 .... x_n
$$
and
$$
\beta_0
$$
denotes the bias term

collecting the weights and input vars, it sums to 
$$
\begin{bmatrix}
\beta_1 & \beta_2 & \cdots & \beta_n
\end{bmatrix}
\quad \text{and} \quad
\begin{bmatrix}
x_1 & x_2 & \cdots & x_n
\end{bmatrix}
$$

fixing that into Z, we get
$$
z = \beta \cdot x^T + \beta_0
$$




Thus, we assume:

$$
\log \left( \frac{p}{1 - p} \right) = z
$$



### Step 2: Solve for P

To express 
$$ p 
$$
in terms of 
$$
z
$$
we exponentiate both sides of the equation:

$$
\frac{p}{1 - p} = e^z
$$

Now, solve for 
$$
p
$$
Multiply both sides by 
$$
1 - p
$$

$$
p = e^z (1 - p)
$$

Distribute
$$
e^z
$$
on the right-hand side:

$$
p = e^z - e^z p
$$

Now, collect all terms involving 
$$
p
$$
on one side of the equation:

$$
p + e^z p = e^z
$$

Factor out 
$$
p
$$
on the left-hand side:

$$
p(1 + e^z) = e^z
$$
Finally, solve for \( p \) by dividing both sides by \( 1 + e^z \):

$$
p = \frac{e^z}{1 + e^z}
$$



### Step 3: Sigmoid Function

Thus, the probability 
$$
p 
$$
can be written as a function of 
$$
z
$$
which is the **sigmoid function**:

$$
p = \frac{1}{1 + e^{-z}}
$$

This function maps the linear combination 
$$
z
$$
to a value between 0 and 1, which represents the probability of the event occurring.

## Conclusion:
The sigmoid function 
$$
\frac{1}{1 + e^{-z}} 
$$
is derived from the assumption that the log-odds of the probability 
$$
p
$$
follow a linear relationship with the predictor variables.


````python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z_values = np.linspace(-15, 15, 100)
p_values = sigmoid(z_values)

plt.plot(z_values, p_values)
plt.title("Sig Fn")
plt.grid(True)
plt.show()
````
